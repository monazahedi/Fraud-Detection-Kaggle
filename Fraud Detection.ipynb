```python
#Removing error warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the data

import pandas as pd
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
#Determine the size of dataset

data.shape
```

```python
#Checking for null values for each feature

data.isnull().sum()
```

```python
#Description of dataset

data.describe()
```

```python
#Feature engineering

import numpy as np

X = data.drop(['Class', 'Time'], axis = 1)
y = data[['Class']]
y = np.ravel(y) #shape of y is converted from (n, 1) to (n, )
```

```python
#Summarize class distribution
#Majority Class: Negative outcome, class 0.
#Minority Class: Positive outcome, class 1.

from collections import Counter
counter = Counter(y)
print(counter)

#Results show that the dataset is imbalanced
```

```python
#Displaying imbalanced dataset by plotting V3 as a function of V2

import matplotlib.pyplot as plt
from numpy import where

row_ix_0 = where(y == 0)[0]
row_ix_1 = where(y == 1)[0]

plt.scatter(X.iloc[row_ix_0]['V2'], X.iloc[row_ix_0]['V3'], label=str(0))
plt.scatter(X.iloc[row_ix_1]['V2'], X.iloc[row_ix_1]['V3'], label=str(1))
plt.xlabel('V2')
plt.ylabel('V3')
plt.legend()
plt.show()
```

```python
#Oversampling of minority class with Synthetic Minority Oversampling Technique (SMOTE). 
#SMOTE is a data augmentation technique used to duplicate examples in the minority class
#Random Undersampling of majority class.


from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline


# define pipeline
over = SMOTE(sampling_strategy=0.1)  #oversample the minority class with SMOTE to about a 1:10 ratio
under = RandomUnderSampler(sampling_strategy=0.25)  #undersample the majority class to achieve about a 1:4 ratio
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)


# transform the dataset
X, y = pipeline.fit_resample(X, y)


# summarize the new class distribution
counter = Counter(y)
print(counter)


# scatter plot of examples by class label
row_ix_0 = where(y == 0)[0]
row_ix_1 = where(y == 1)[0]

plt.scatter(X.iloc[row_ix_0]['V2'], X.iloc[row_ix_0]['V3'], label=str(0))
plt.scatter(X.iloc[row_ix_1]['V2'], X.iloc[row_ix_1]['V3'], label=str(1))
plt.xlabel('V2')
plt.ylabel('V3')
plt.legend()
plt.show()
```

```python
#Splitting data into training/validation & test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=48)
```

```python
#Detecting the most important features using 80% data for training/validation and 20% for testing

from sklearn.feature_selection import mutual_info_classif
import seaborn as sns

mutual_information = mutual_info_classif(X_train, y_train, n_neighbors=5, copy = True)
plt.subplots(1, figsize=(28, 1))
sns.heatmap(mutual_information[:, np.newaxis].T, cmap='Blues', cbar=False, linewidths=1, annot=True, annot_kws={"size": 12})
plt.yticks([], [])
plt.gca().set_xticklabels(X.columns, rotation=45, ha='right', fontsize=12)
plt.suptitle("Variable Importance (mutual_info_classif)", fontsize=12, y=1.2)
plt.gcf().subplots_adjust(wspace=0.4)
```

```python
#Restricting features to 'V2','V3','V4','V7','V9','V10','V11','V12','V14','V16','V17','V18','V21','V27','Amount'

X_opt = X[['V2','V3','V4','V7','V9','V10','V11','V12','V14','V16','V17','V18','V21','V27','Amount']]
```

```python
#Splitting data into training/validation & test sets

X_train, X_test, y_train, y_test = train_test_split(X_opt, y, test_size=0.3, random_state=48)
```

```python
#Standardizing features

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#XGBoost
#Determining optimum hyperparameters using the train/validation set 
#Gridsearch with 5-fold cross-validation 

import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline


model = xgb.XGBClassifier(use_label_encoder=False, random_state=48)

pipeline = Pipeline([  
    ('model', model)
])


param_grid={'model__n_estimators':[500],
        'model__max_depth':[5],
        'model__eta':[0.3], 
        'model__min_child_weight':[5], 
        'model__reg_alpha':[0.2], 
        'model__reg_lambda':[1], 
        'model__gamma':[0]}


#The ROC_AUC score is great for imbalanced datasets
grid = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, verbose=0, scoring='roc_auc')

grid.fit(X_train, y_train)
```

```python
#Reporting the cross-validation score and optimum hyperparameters

print("Results from Grid Search")
print("\n The best score across all searched params:\n",'{:.2%}'.format(grid.best_score_))
print("\n The best parameters across all searched params:\n",grid.best_params_)
```

```python
#Determining predictions on unseen test set

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

pred = grid.predict(X_test)
acc_score = '{:.2%}'.format(accuracy_score(y_test,pred))
pre_score = '{:.2%}'.format(precision_score(y_test,pred))
rec_score = '{:.2%}'.format(recall_score(y_test,pred))
f1_score = '{:.2%}'.format(f1_score(y_test,pred))
roc_auc_score = '{:.2%}'.format(roc_auc_score(y_test,pred))

print("Results on Test Set of XGBoost Classifier")
print("\n Accuracy Score:\n",acc_score)
print("\n Precision Score:\n",pre_score)
print("\n Recall Score:\n",rec_score)
print("\n F1 Score:\n",f1_score)
print("\n ROC-AUC Score:\n",roc_auc_score)
```

```python
#Plotting the confusion matrix

from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(grid, X_test, y_test)  
plt.show()
```
